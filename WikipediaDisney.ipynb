```python
def clean_data(something):
    for sup in something.find_all("sup"):
        sup.decompose()

def clean_data2(soup):
    for span in soup.find(class_='infobox vevent'):
        spans = span.find_all("span")
        for span in spans:
            span.decompose()

def find_row(row_data):
        if row_data.find("li") == True:
            return [li.get_text(", ",strip=True) for li in row_data.find_all("li")]
        else:
            return row_data.get_text(", ",strip=True)

def get_content(url):
    r = requests.get(url)
    soup = bs(r.content,"lxml")

    info_box = soup.find(class_='infobox vevent')
    trs = info_box.find_all('tr')

    clean_data(soup)
    clean_data2(soup)

    movie_info = {}
    for index, tr in enumerate(trs):
        if index == 0:
            movie_info['title'] = tr.get_text(strip=True)
        elif index == 1:
            continue
        else:
            label = tr.find("th").get_text(" ",strip=True)
            data = find_row(tr.find("td"))
            movie_info[label] = data.replace("\xa0", " ").split(",")
```

```python
def clean_tag(something):
    for sup in something.find_all("sup"):
        sup.decompose()


def find_row(row_data):
        if row_data.find("li") == True:
            return [li.get_text(", ",strip=True) for li in row_data.find_all("li")]
        else:
            return row_data.get_text(", ",strip=True)

def get_content(url):
    r = requests.get(url)
    soup = bs(r.content,"lxml")

    info_box = soup.find(class_='infobox vevent')
    trs = info_box.find_all('tr')

    clean_tag(soup)

    movie_info = {}
    for index, tr in enumerate(trs):
        if index == 0:
            movie_info['title'] = tr.get_text(strip=True)
        elif index == 1:
            continue
        else:
            label = tr.find("th").get_text(" ",strip=True)
            data = find_row(tr.find("td"))
            clean_tag(tr.find("td"))
            movie_info[label] = data.replace("\xa0", " ").split(",")
    return movie_info
```

```python
movies = soup.select(".wikitable.sortable i a")
base_path = "https://en.wikipedia.org"

movie_info_list = []
for index, movie in enumerate(movies):
    try:
        relative_path = movie['href']
        path = base_path + relative_path
        title = movie['title']

        movie_info_list.append(get_content(path))
    except Exception as e:
        print(movie.get_text())
        print(e)
    
```

```python
import json

def save_data(file_name,data):
    with open(file_name, 'w', encoding='utf-8') as f:
        return json.dump(data, f, ensure_ascii=False, indent=2)

def load_data(file_name):
    with open(file_name, encoding='utf-8') as f:
        return json.load(f)
        
```
